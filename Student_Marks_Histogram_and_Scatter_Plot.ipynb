```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Case_Study_5_B.xlsx")

df['Percentage'] = (df['Marks'] / 15) * 100


high_achievers = df[df['Percentage'] > 75]
average_students = df[(df['Percentage'] >= 60) & (df['Percentage'] <= 75)]
low_achievers = df[df['Percentage'] < 60]

print("Students with greater than 75% marks:")
print(high_achievers)
print("\nStudents with marks between 60% and 75%:")
print(average_students)
print("\nStudents with less than 60% marks:")
print(low_achievers)

plt.figure(figsize=(10, 6))
plt.hist([high_achievers['Percentage'], average_students['Percentage'], low_achievers['Percentage']], bins=range(0, 101, 5), color=['blue', 'green', 'red'], edgecolor='black', label=['>75%', '60-75%', '<60%'])

plt.xlabel("Percentage")
plt.ylabel("Number of Students")
plt.title("Histogram Plot")
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(high_achievers.index, high_achievers['Percentage'], label='>75%', color='pink')
plt.scatter(average_students.index, average_students['Percentage'], label='60-75%', color='green')
plt.scatter(low_achievers.index, low_achievers['Percentage'], label='<60%', color='red')
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```
